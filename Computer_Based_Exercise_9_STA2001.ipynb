```python
import matplotlib.pyplot as plt
from matplotlib import animation
import math
import numpy as np
import random
import scipy.stats as st
import seaborn as sns
```

```python
N = 100
Y = np.random.uniform(size=N)
Y
X = -(np.log(1-Y))
G = []

x = np.linspace(0, 10,1000)
for i in x:
  counter = 0
  for j in X:
    if j<i:
      counter+=1
  G.append(counter/100)

F = 1-np.exp(-x)

fig,ax= plt.subplots()
plt.style.use('ggplot')
ax.plot(x,F, label="F(x)= 1-exp(-x)")
ax.plot(x,G, label="G(x) = N(x)/100")
ax.set_title('100 Realizations')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.legend()
```
